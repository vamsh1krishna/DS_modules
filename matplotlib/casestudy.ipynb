```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)')
plt.plot(gas.Year, gas.USA,label='United States')
plt.plot(gas.Year, gas.Canada,label='Canada')
#can also call year in different way
plt.plot(gas['Year'], gas['South Korea'],label='South Korea')

plt.legend()
plt.xticks(gas.Year)

plt.show()
```

```python
#to declutter the xaxis, we can run iterations of years from file with intervals
#with a list of range b, list elements can iterated at regular intervals of x (3 in this case) through List[::x]
print(gas.Year[::3])
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)')
plt.plot(gas.Year, gas.USA,'c.-',label='United States')
plt.plot(gas.Year, gas.Canada,'k-',label='Canada')
#can also call year in different way
plt.plot(gas['Year'], gas['South Korea'],'y.-',label='South Korea')
plt.plot(gas.Year,gas.Australia,'r.-',label='Australia')

plt.legend()
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Prices in USD')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

plt.pie([left,right])
plt.show()
# labels
```

```python
#labelling and coloring them, autopercantage'%.2f'
labels=['Left','Right']
colors =['#42f56f','#e0d71d']
#to show percent symbol on pie chart '%.2f%%'
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f%%')
plt.title('Foot Prefernece of FIFA Players')
plt.show()
```

```python
fifa.Weight
##as weight is pounds and "lbs" string is at the end we need to remove the string and use the weight
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x) ==str else x for x in fifa.Weight]
fifa.Weight
```

```python
#categorise players based on weight
fifa.Weight
light =fifa.loc[fifa.Weight<125].count()[0]
light_medium =fifa.loc[(fifa.Weight>=125) &(fifa.Weight<150) ].count()[0]
medium =fifa[(fifa.Weight>=150) &(fifa.Weight<175) ].count()[0]
medium_heavy =fifa[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy =fifa[fifa.Weight>=200].count()[0]

##if we dont waant default colors, then we can use plt.style.use('ggplot')
plt.style.use('ggplot')

##Plot
weights=[light,light_medium,medium,medium_heavy,heavy]
labels =['under 125','125-150','150-175','175-200','Over 200']
plt.pie(weights,labels=labels,autopct='%.2f %%')

plt.show()





```

```python
##Plot
weights=[light,light_medium,medium,medium_heavy,heavy]
labels =['under 125','125-150','150-175','175-200','Over 200']
##as percentages are getting overlapped , we can use pct distance
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=2)

plt.show()
```

```python
##Plot
weights=[light,light_medium,medium,medium_heavy,heavy]
labels =['under 125','125-150','150-175','175-200','Over 200']
##explode property, this makes pie chart split up
##as under 125 and over 200 are still very close we can use different value to splitup differently, 0 = doesnt split
explode=(.4,.2,0,0,.4)
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=2,explode=explode)
plt.title('Weight Distribution of Players in (Lbs)')

plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,12))
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels=['FC barcelona', 'Real Madrid','NE Revolution']
plt.boxplot([barcelona,Madrid,revs],labels=labels)


plt.title('Comparision among FC BArcelona, Real Madrid and New England Revolution'  )
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
## we can color boxes in plt.boxplot, but all boxes will get same color
##to give different colors to different boxes, we have to mention patch_artist=True, before applying face colors
## median lone thickness can also be adjusted with medianprops={}
plt.style.use('default')
plt.figure(figsize=(5,12))
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels=['FC barcelona', 'Real Madrid','NE Revolution']
boxes = plt.boxplot([barcelona,Madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})
fill=['#e0a61d','#1de0da','#bde01d']
for box in boxes['boxes']:
    #set edge color
    box.set(color = '#4286f4', linewidth=2)

    #change Fill color
    box.set(facecolor=fill.pop(0))


plt.title('Comparision among FC BArcelona, Real Madrid and New England Revolution'  )
plt.ylabel('FIFA Overall Rating')
plt.show()
```
