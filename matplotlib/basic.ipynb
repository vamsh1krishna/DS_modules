Load necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic graph

```python
x=[1,2,3]
y=[3,6,9]
plt.plot(x,y)
plt.title('Our First Graph!')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

title and labels font and size  cam be added "matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)"

```python
x=[1,2,3]
y=[3,6,9]
plt.plot(x,y)
plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')
plt.show()
```

to give numbers on axis (like integers) and graph size adjusts automatically

```python
x=[0,1,2,3]
y=[0,3,6,9]
plt.plot(x,y)
plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,3,6,9,12])
plt.show()
```

marking down the grapgh line(legand). for that legand has to be defined during plot. coloring also can be given like red (or) r, yellow (or) y
naming the legand, color and linewidth are parameters of plot and many mor eparameters

```python
x=[0,1,2,3,4]
y=[0,3,6,9,12]
plt.plot(x,y, label='3x', color='r', linewidth=2, linestyle = '--', marker='.',markersize = 10 )

plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,3,6,9,12,15])


plt.show()
```

use shorthand technique for defining the graph using #fmt
#fmt = '[color][marker][line]'

```python
x=[0,1,2,3,4]
y=[0,3,6,9,12]

#plt.plot(x,y, label='3x', color='r', linewidth=2, linestyle = '--', marker='.',markersize = 10 )
#use shorthand technique for defining the graph using #fmt
#fmt = '[color][marker][line]'
plt.plot(x,y, 'r.--',label='3x')

plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,3,6,9,12,15])

plt.legend()

plt.show()
```

second line in graph

```python
x=[0,1,2,3,4]
y=[0,3,6,9,12]

#plt.plot(x,y, label='3x', color='r', linewidth=2, linestyle = '--', marker='.',markersize = 10 )
#use shorthand technique for defining the graph using #fmt
#fmt = '[color][marker][line]'
plt.plot(x,y, 'r.--',label='3x')

##line number two
x2 = np.arange(0,4,0.5)
plt.plot(x2, x2**2, 'k*-.',label='x^2')


plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

#plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,3,6,9,12,15])

plt.legend()

```

#can limit the number of values in each plot graph, when required as graph overlaps. such as second line in below example

```python
x=[0,1,2,3,4]
y=[0,3,6,9,12]

#plt.plot(x,y, label='3x', color='r', linewidth=2, linestyle = '--', marker='.',markersize = 10 )
#use shorthand technique for defining the graph using #fmt
#fmt = '[color][marker][line]'
plt.plot(x,y, 'r.--',label='3x')

##line number two
x2 = np.arange(0,5,0.5)
#plot the selected part of graph as line
plt.plot(x2[:5], x2[:5]**2, 'k*-',label='x^2')
#plot the remaining part of graph as dotted line
plt.plot(x2[:8], x2[:8]**2, 'b.-.',label='x^2')

plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

#plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,3,6,9,12,15])

plt.legend()


```

resizeing the graph.(dpi of 300 is recommended)

```python
x=[0,1,2,3,4]
y=[0,3,6,9,12]

plt.figure(figsize=(5,3),dpi = 300)

plt.plot(x,y, 'r.--',label='3x')

##line number two
x2 = np.arange(0,5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'k*-',label='x^2')
plt.plot(x2[:8], x2[:8]**2, 'b.-.',label='x^2')

plt.title('Our First Graph!',fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')



plt.legend()

plt.show()
```

### Bar Chart

```python
labels = ['A','B','c']
values =[1,4,2]
#can add titles,colors same way as linegraph 
plt.bar(labels,values)

plt.figure(figsize=(6,4))

plt.show()
```

```python
#to make patterns on bar
labels = ['A','B','c']
values =[1,4,2]
#can add titles,colors same way as linegraph 
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')



plt.figure(figsize=(6,4))

plt.show()
```

```python
#to make patterns on bar
labels = ['A','B','c']
values =[1,4,2]
#can add titles,colors same way as linegraph 
bars = plt.bar(labels,values)
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')
patterns =['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))



plt.figure(figsize=(6,4))

plt.show()
```
