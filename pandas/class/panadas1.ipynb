Filtering the column by one or more criteria

```python
shows[shows['country']=='India'].head(33)
```

```python
by_india = shows[shows['country']=='India']
movies_by_india = by_india[by_india['type']=='Movie']
movies_by_india.head(3)
```

```python
shows_by_india = shows['country']=='India'
movies = shows['type']=='Movie'
shows[shows_by_india & movies].head(3)
```

```python
#Indian movies before 2003
before_2000 = shows['release_year']<= 2003
shows_by_india = shows['country']=='India'
movies = shows['type']=='Movie'
shows[shows_by_india & movies & before_2000].head(3)
```

```python
#Indian movies between 2010 and 2020
between_2010_and_2020 = shows['release_year'].between(2010,2020)
movies = shows['type']=='Movie'
shows[movies & between_2010_and_2020].head(3)
```

```python
shows['duration']
```

```python
#remove min/hours/seasons from duration column
shows["duration"].dropna().str.replace(" min","",regex=False).str.replace(" Seasons","",regex=False).str.replace(" Season","",regex=False).astype(int)
```

```python
#Find all the american movies released in the year 2021
year_2021 = shows['release_year']== 2021
shows_by_america = shows['country']=='United States'
movies = shows['type']=='Movie'
shows[shows_by_america & movies & year_2021].head(3)

#Find all the american tv shows 
shows_by_america = shows['country']=='United States'
tv_shows = shows['type']=='TV Show'
shows[shows_by_america & tv_shows ].head(3)
```
