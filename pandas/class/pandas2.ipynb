Series Object
-------------
Series is One-dimensional labelled array for homogenous data.

```python
import pandas as pd
import numpy as np
```

```python
pd.Series()
```

```python
icecream_flavors = ["vanilla","Butterscotch","Strawberry","Chocolate"]
pd.Series(icecream_flavors)
```

```python
pd.Series(
    data = None,
    index=None,
    dtype=None,
    name=None,
    name=None,
    copy=False,
    fastpath=None
)
```

```python
icecream_flavors = ["vanilla","Butterscotch","Strawberry","Chocolate"]
days_of_week = ("Monday","Wednesday","Thursday","Saturday")
pd.Series(icecream_flavors,index=days_of_week)
```

```python
#create a Series oobject with gold prices and days_of_week
days_of_week = ("Monday","Wednesday","Thursday","Saturday")
prices_of_week=[46600,46700,46800,46900]
pd.Series(data=prices_of_week,index=days_of_week)
```

```python
# Fill NaN values
netflix_titles["country"].fillna('United States',inplace=True)
netflix_titles.tail(6)
```

```python
netflix_titles = pd.read_csv(r'netflix_titles.csv',parse_dates=['date_added'])
netflix_titles.nunique()
```

```python
#lets categorize the data
netflix_titles['type'].astype('category')
```

```python
#filter by release year
netflix_titles[netflix_titles['release_year']==2021]
```

```python
#inversion with ~
my_series= pd.Series([True,False,True])
print(my_series)
print('inverted_series')
print(~my_series)

#lets apply this
is_indian = netflix_titles['country']=="India"
netflix_titles[~is_indian].head(3)
```

```python
# isin
is_american = netflix_titles['country']=="United States"
is_indian = netflix_titles['country']=="India"
print(netflix_titles[is_american|is_indian].head())
print('#'*100)
countries = ["India","United States"]
is_titles_in_countries = netflix_titles['country'].isin(countries)
netflix_titles[is_titles_in_countries]
```

```python
#between
#write an expression to find all the titles between and 1990 and 2000
post_1990 = netflix_titles['release_year']>1990  
before_2000 = netflix_titles['release_year']<2000
# netflix_titles[post_1990 & before_2000].head(10)
between_1999_and_2000 = netflix_titles['release_year'].between(left=1990,right=2000)
netflix_titles[between_1999_and_2000].head()
```

```python
#Find the null values in countries
#Find countries,cast and directors which are not null
is_country_null = netflix_titles['country'].isnull()
is_cast_null = netflix_titles['cast'].isnull()
director_not_null = netflix_titles['director'].notnull()
netflix_titles[~is_country_null & ~is_cast_null & director_not_null].head()
```

```python
#Dealing with duplicates
is_duplicated = netflix_titles['country'].duplicated()
netflix_titles[is_duplicated]

#drop duplicates in listed_in
netflix_titles.drop_duplicates(subset=['listed_in'],keep='first')

#drop duplicates in description
netflix_titles.drop_duplicates(subset=['description'],keep='last')

#drop duplicates in date_added
netflix_titles.drop_duplicates(subset=['date_added'],keep='last')

#drop duplicates with titles
netflix_titles.drop_duplicates(subset=['title'],keep='first')
```
