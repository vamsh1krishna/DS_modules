## Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels =['Fruits','Vegetables','Others']
counts =[3,5,1]
plt.bar(labels,counts)
plt.show()
```

```python
food_items_1 =[1,1] #1 of fruits and 1 of vegetables
food_items_2 =[1,2] #1 of fruits and 2 of vegetables
food_items_3 =[2,3] #2 of fruits and 3 of vegetables

counts = [food_items_1,food_items_2,food_items_3]
locations = np.array([0,1,2])
labels =["Fruits","Vegetables"]
width = 0.3
bars_fruits = plt.bar(locations ,[food_item[0] for food_item in counts])
bars_vegetables = plt.bar(locations ,[food_item[1] for food_item in counts],bottom=bars_fruits)
# bars_others = plt.bar(locations ,[food_item[2] for food_item in counts])

plt.xticks(locations,['Fruits','Vegetables','others'])
plt.legend([bars_fruits,bars_vegetables],labels)
plt.show()
```

```python
## Visualizations of probability Distributions
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
samples = np.random.normal(0,1,size=1000)
x = np.linspace(samples.min(),samples.max(),1000)
y = stats.norm.pdf(x)
plt.hist(samples,alpha=0.25,bins=20,density=True)
plt.plot(x,y)
plt.show()
```

## Visualizations shorthand from Seaborn and pandas

```python
import pandas as pd
x = np.random.normal(0,1,1000)
y = np.random.normal(5,2,1000)
df = pd.DataFrame({'column_1':x,'column_2':y})
df.head()
```

```python
import seaborn as sns
sns.jointplot(x='column_1',y='column_2',data=df)
plt.show()
```
